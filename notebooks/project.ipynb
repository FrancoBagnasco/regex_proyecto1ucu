# Proyecto 1 - RegEx


```python
import re

# Lista de archivos HTML de donde extraeremos los datos
files = [
    '../data/baby1990.html',
    '../data/baby1992.html',
    '../data/baby1994.html',
    '../data/baby1996.html',
    '../data/baby1998.html',
    '../data/baby2000.html',
    '../data/baby2002.html',
    '../data/baby2004.html',
    '../data/baby2006.html',
    '../data/baby2008.html',
]

# Bucle para que el código lea cada archivo HTML 
for file in files:
    with open(file, "r") as file:
        all_files = file.read()

    # Encuentra y extrae el año de donde se extraeran los datos
    year_match = re.search(r'Popularity in (\d{4})', all_files)
    if year_match:
        year = year_match.group(1)
        
    # Encuentra y extrae los nombres con sus respectivas posiciones del ranking
    names_matches = re.findall(r'<td>(\d+)</td><td>(\w+)</td><td>(\w+)</td>', all_files)

    # Almacena los nombres y sus posiciones en un diccionario
    names_dict = {}
    for rank, boy_name, girl_name in names_matches:
            names_dict[boy_name] = rank
            names_dict[girl_name] = rank
            
    # Crea una lista con los nombres ordenados alfabéticamente
    sorted_names = sorted(names_dict.keys())
    
    # Crea una lista con el formato requerido
    result_list = [f'--------------------------------------------------------\nPara el año {year} el ranking de nombres fue el siguiente:\n--------------------------------------------------------']
    for name in sorted_names:
        result_list.append(f'{name}: #{names_dict[name]}')

    # Imprime los resultados
    for item in result_list:
        print(item)
```
